**BASIC INDEXING**

```python
x = torch.tensor([10, 20, 30, 40, 50])

print(x[0])     # First element
print(x[-1])    # Last element
```

Slicing (1D Tensor)

```python
print(x[1:4])     # From index 1 to 3
print(x[:3])      # First 3 elements
print(x[::2])     # with Step size = 2
```

```python
mat[1,2]
```

Slicing in 2D Tensors

```python
#torch.device('cuda') is a standard way in the PyTorch deep learning framework to specify that a tensor or model should be
#allocated and run on the GPU (Graphics Processing Unit), assuming one is available and properly configured

#This indicates that PyTorch is utilizing NVIDIA's CUDA platform to leverage the parallel
# processing power of the GPU for faster computations.
device = torch.device('cuda')
```
